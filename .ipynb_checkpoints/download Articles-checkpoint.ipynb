```python
import selenium
import requests
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
chrome_options = Options()
chrome_options.add_argument("--headless")
```

```python
driver = webdriver.Chrome(options=chrome_options)
```

```python
nip2019 = "https://papers.nips.cc/book/advances-in-neural-information-processing-systems-32-2019"
```

```python
driver.get(nip2019)
```

```python
ulElements = driver.find_elements_by_tag_name("ul")
```

```python
ul = ulElements[1]
```

```python
liElemetns = ul.find_elements_by_tag_name("li")
```

```python
for li in liElemetns:
    aElements = li.find_element_by_tag_name("a")
    href = aElements.get_attribute('href')
    print(href)
```

```python
def downloadPaperData(url):
    driver.get(href)
    title = driver.find_element_by_class_name("subtitle").text
    abstract = driver.find_element_by_class_name("abstract").text
    authors = getAuthors()
    
    
```

```python
def getAuthors():
    ulElements = driver.find_elements_by_tag_name("ul")
    ul = ulElements[1]
    
    liElemetns = ul.find_elements_by_tag_name("li")
    
    authors = ""
    for li in liElemetns:
        if len(authors) == 0:
            authors = li.text
        else:
            authors += ":" + li.text
    return authors
```

```python
def downloadPdf(did):
    aTags = driver.find_elements_by_tag_name("a")
    for a in aTags:
        if a.text == "[PDF]":
            url = a.get_attribute('href')
            
    response = requests.get(url)
    with open('/hdd/dataPhdProject/{}.pdf'.format(did), 'wb') as f:
        f.write(response.content)
```
